```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Format 'Date' column as a string
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

# Split 'ShippingAddress' into house number and street name
# Assuming address format is 'NUMBER StreetName'
df['ShippingAddressNumber'] = df['ShippingAddress'].apply(lambda x: int(x.split(' ')[0]))
df['ShippingStreetName'] = df['ShippingAddress'].apply(lambda x: ' '.join(x.split(' ')[1:]))


# Drop original 'ShippingAddress' column as requested
df = df.drop(columns=['ShippingAddress'])

# Reorder columns for better readability
desired_order = [
    'OrderID', 'Date', 'CustomerID', 'Product',
    'Quantity', 'UnitPrice', 'ShippingAddressNumber', 'ShippingStreetName',
    'PaymentMethod', 'OrderStatus', 'TrackingNumber', 'ItemsInCart',
    'CouponCode', 'ReferralSource', 'TotalPrice'
]

# Add any other existing columns not explicitly mentioned in desired_order to the end
existing_cols = df.columns.tolist()
final_order = [col for col in desired_order if col in existing_cols] + \
              [col for col in existing_cols if col not in desired_order]
df = df[final_order]

print("DataFrame after transformations and reordering:")
display(df.head())
display(df.info())
```

```python
# List the columns that should be categories
category_cols = ['PaymentMethod', 'OrderStatus', 'ReferralSource']

# Convert them to the category data type
df[category_cols] = df[category_cols].astype('category')

# Optional: Run a quick check to confirm the types have updated
df.info()
```

```python
from IPython.display import display, HTML

results = []
for col in df.columns:
        unique_vals = df[col].unique()
        num_unique = len(unique_vals)
        # Limit the display of unique values to avoid excessively long outputs
        if num_unique > 10:
            display_vals = ', '.join(map(str, unique_vals[:5])) + ', ...'
        else:
            display_vals = ', '.join(map(str, unique_vals))
        results.append({'Column': col, 'Number of Unique Values': num_unique, 'Unique Values (Sample/All)': display_vals})

unique_df = pd.DataFrame(results)
display(HTML(unique_df.to_html(index=False)))
```

```python
# 1. Dates: Lock into ISO 8601 Format (YYYY-MM-DD)
# This forces the datetime object into a strict text format
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')

# 2. Text: Trim whitespace and apply Proper Case (Title Case)
# We will loop through your main text categories
text_cols = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource', 'ShippingStreetName']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# If you converted these to categories earlier, you can lock them back into categories:
category_cols = ['PaymentMethod', 'OrderStatus', 'ReferralSource']
df[category_cols] = df[category_cols].astype('category')

# 3. Numbers: Numeric Precision to exactly 2 decimal places
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)
```

```python
from IPython.display import display, HTML

results = []
for col in df.columns:
        unique_vals = df[col].unique()
        num_unique = len(unique_vals)
        # Limit the display of unique values to avoid excessively long outputs
        if num_unique > 10:
            display_vals = ', '.join(map(str, unique_vals[:5])) + ', ...'
        else:
            display_vals = ', '.join(map(str, unique_vals))
        results.append({'Column': col, 'Number of Unique Values': num_unique, 'Unique Values (Sample/All)': display_vals})

unique_df = pd.DataFrame(results)
display(HTML(unique_df.to_html(index=False)))
```
